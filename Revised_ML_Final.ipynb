```python
import fastf1
import pandas as pd

#fastf1.Cache.enable_cache('cache')  # speeds up repeated runs

street_circuits = ["Monaco Grand Prix", "Azerbaijan Grand Prix", "Las Vegas Grand Prix"]

all_laps = []
race_rounds = list(range(1, 25))

for rnd in race_rounds:
    print(f"Loading race {rnd}...")
    session = fastf1.get_session(2024, rnd, 'R')
    session.load(telemetry=False, weather=True)

    laps = session.laps.copy()

    # Keep Test 1 columns
    laps = laps[["Driver", "LapNumber", "LapTime", "Compound", "Team"]].copy()
    laps.rename(columns={"Compound": "TireCompound"}, inplace=True)

    # Circuit type
    laps["CircuitType"] = "Permanent"
    if session.event["EventName"] in street_circuits:
        laps["CircuitType"] = "Street"

    # Tire age
    laps["TireAge"] = laps.groupby("Driver")["LapNumber"].cumcount() + 1

    # Rainfall (boolean)
    if hasattr(session, "weather") and session.weather is not None:
        laps["Rainfall"] = int(session.weather['Rainfall'].max() > 0)
    else:
        laps["Rainfall"] = 0

    # Lap time in seconds
    laps["LapTimeSeconds"] = laps["LapTime"].dt.total_seconds()

    # Race name
    laps["RaceName"] = session.event["EventName"]

    # Avg speed
    circuit_len_km = track_lengths_km[session.event["EventName"]]
    laps["AvgSpeed"] = circuit_len_km / (laps["LapTimeSeconds"] / 3600)

    all_laps.append(laps)

# Combine all races
df_laps = pd.concat(all_laps, ignore_index=True)

# Split into train/test
train_df = df_laps[df_laps["RaceName"] != "Abu Dhabi Grand Prix"].copy()
test_df = df_laps[df_laps["RaceName"] == "Abu Dhabi Grand Prix"].copy()

print("Done! Train shape:", train_df.shape, "Test shape:", test_df.shape)
train_df.head()
```

```python
train_df = df_laps[df_laps["RaceName"] != "Abu Dhabi Grand Prix"]
test_df = df_laps[df_laps["RaceName"] == "Abu Dhabi Grand Prix"]
```

```python
from sklearn.preprocessing import OneHotEncoder

# Features
categorical_features = ["Driver", "Team", "TireCompound", "CircuitType"]
numeric_features = ["LapNumber", "TireAge", "Rainfall"]
target = ["LapTimeSeconds"]  # or "AvgSpeed"

# Remove rows with missing target
train_df = train_df.dropna(subset=target)
test_df = test_df.dropna(subset=target)

# Split features and target
X_train = train_df[categorical_features + numeric_features].copy()
y_train = train_df[target].copy()
X_test = test_df[categorical_features + numeric_features].copy()
y_test = test_df[target].copy()

# One-hot encode categorical features
encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
X_train_cat = encoder.fit_transform(X_train[categorical_features])
X_test_cat = encoder.transform(X_test[categorical_features])

# Convert to DataFrame and combine with numeric features
import pandas as pd

X_train_final = pd.concat(
    [X_train[numeric_features], pd.DataFrame(X_train_cat, columns=encoder.get_feature_names_out(categorical_features), index=X_train.index)],
    axis=1
)
X_test_final = pd.concat(
    [X_test[numeric_features], pd.DataFrame(X_test_cat, columns=encoder.get_feature_names_out(categorical_features), index=X_test.index)],
    axis=1
)

print("X_train_final shape:", X_train_final.shape)
print("X_test_final shape:", X_test_final.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

# --- Train model ---
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train_final, y_train.values.ravel())

# --- Predict on test set ---
y_pred = model.predict(X_test_final)

# --- Build prediction dataframe ---
pred_df = test_df.copy()
pred_df["PredLapTime"] = y_pred

# --- Compute per-driver predicted vs actual averages ---
avg_preds = pred_df.groupby("Driver")["PredLapTime"].mean()
avg_actuals = pred_df.groupby("Driver")["LapTimeSeconds"].mean()

# --- Comparison dataframe ---
comparison = pd.DataFrame({
    "Predicted": avg_preds,
    "Actual": avg_actuals
})

comparison["Error"] = comparison["Predicted"] - comparison["Actual"]

print(comparison)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(comparison["Actual"], comparison["Predicted"])
rmse = mse ** 0.5
mae = mean_absolute_error(comparison["Actual"], comparison["Predicted"])
r2 = r2_score(comparison["Actual"], comparison["Predicted"])

print("Linear Regression Performance (Driver Averages):")
print(f"RMSE: {rmse:.3f} sec")
print(f"MAE: {mae:.3f} sec")
print(f"R²: {r2:.3f}")


print("Driver-wise Comparison:\n", comparison)


plt.figure(figsize=(12,6))
sns.barplot(x=comparison.index, y=comparison["Predicted"], color='yellow', label="Predicted")
sns.barplot(x=comparison.index, y=comparison["Actual"], color='green', alpha=0.5, label="Actual")
plt.ylabel("Average Lap Time (sec)")
plt.title("Predicted vs Actual Average Lap Time per Driver — Abu Dhabi GP")
plt.xticks(rotation=45)
plt.legend()
plt.show()

#Error displays delta between Predicted Time and Actual Time.
plt.figure(figsize=(12,5))
sns.barplot(x=comparison.index, y=comparison["Error"], palette="coolwarm")
plt.axhline(0, color='black', linestyle='--')
plt.ylabel("Prediction Error (sec)")
plt.title("Prediction Error per Driver — Linear Regression")
plt.xticks(rotation=45)
plt.show()
```
